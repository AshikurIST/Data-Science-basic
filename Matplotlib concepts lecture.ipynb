```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
# Functional methode 

plt.plot(x,y,'r-')
plt.xlabel('X lable')
plt.ylabel('Y lable')
plt.title('TITLE')
```

```python
# multiple plot 

plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# object oriented methode

fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)

axes.set_xlabel('x label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])    #add_axes([left,bottom,width,hight])
axes1.set_title('title 1')
axes1.plot(x,y)
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')


axes2=fig.add_axes([0.2,0.39,0.32,0.25]) 
axes2.plot(y,x)
axes2.set_title('title 2')
```
